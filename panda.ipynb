## Income Distribution by Age Group

### Subtask:
Create a bar chart to visualize the distribution of income (income_binary) across different age groups. This will show how income levels vary between youth, adult, and senior categories. Ensure legends are provided.


**Reasoning**:
First, I need to import the necessary libraries, `matplotlib.pyplot` and `seaborn`, to create the bar chart.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average income_binary for each age_group
average_income_by_age_group = df.groupby('age_group')['income_binary'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='income_binary', data=average_income_by_age_group, palette='viridis')

# Set labels and title
plt.xlabel('Age Group')
plt.ylabel('Average Income (Binary)')
plt.title('Income Distribution by Age Group')

# Display the plot
plt.show()
```

**Reasoning**:
The previous code generated `FutureWarning` messages. I will fix the first warning by adding `observed=False` to the `groupby` operation to explicitly handle categorical data. For the second warning, I will follow the suggestion to assign the `x` variable (`age_group`) to `hue` and set `legend=False` in the `sns.barplot` call to prevent its deprecation and ensure proper coloring without a redundant legend.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average income_binary for each age_group
average_income_by_age_group = df.groupby('age_group', observed=False)['income_binary'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='income_binary', data=average_income_by_age_group, hue='age_group', palette='viridis', legend=False)

# Set labels and title
plt.xlabel('Age Group')
plt.ylabel('Average Income (Binary)')
plt.title('Income Distribution by Age Group')

# Display the plot
plt.show()
```

## Average Income by Education Level

### Subtask:
Create a bar chart to visualize the average `income_binary` for each education level. This chart will help understand how income levels correlate with different educational backgrounds. Ensure legends are provided if applicable.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average income_binary for each education level
average_income_by_education = df.groupby('education')['income_binary'].mean().reset_index()

# Sort by average income for better visualization
average_income_by_education = average_income_by_education.sort_values(by='income_binary', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='income_binary', y='education', data=average_income_by_education, palette='crest')

# Set labels and title
plt.xlabel('Average Income (Binary)')
plt.ylabel('Education Level')
plt.title('Average Income by Education Level')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average income_binary for each education level
average_income_by_education = df.groupby('education')['income_binary'].mean().reset_index()

# Sort by average income for better visualization
average_income_by_education = average_income_by_education.sort_values(by='income_binary', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='income_binary', y='education', data=average_income_by_education, hue='education', palette='crest', legend=False)

# Set labels and title
plt.xlabel('Average Income (Binary)')
plt.ylabel('Education Level')
plt.title('Average Income by Education Level')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for workclass distribution
plt.figure(figsize=(12, 7))
sns.countplot(y='workclass', data=df, order=df['workclass'].value_counts().index, palette='magma')

# Set labels and title
plt.xlabel('Count')
plt.ylabel('Workclass')
plt.title('Distribution of Individuals by Workclass')

# Display the plot
plt.show()
```

**Reasoning**:

## Relationship vs. Capital Gain

### Subtask:
Generate a bar chart showing the average 'capital-gain' for each 'relationship' status. This will indicate which relationship statuses tend to have higher capital gains on average.


**Reasoning**:
I need to calculate the average capital-gain for each relationship status and then create a bar chart to visualize this relationship, as specified in the subtask.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the average capital-gain for each relationship status
average_capital_gain_by_relationship = df.groupby('relationship')['capital-gain'].mean().reset_index()

# Sort by average capital gain for better visualization
average_capital_gain_by_relationship = average_capital_gain_by_relationship.sort_values(by='capital-gain', ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='relationship', y='capital-gain', data=average_capital_gain_by_relationship, hue='relationship', palette='plasma', legend=False)

# Set labels and title
plt.xlabel('Relationship Status')
plt.ylabel('Average Capital Gain')
plt.title('Average Capital Gain by Relationship Status')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```
